```python
# _*_ logging: try/exception practices_*_ #

# _*_ traceback.print_exc(file=open(filename, mode='a', encoding='utf-8')) _*_ #

import logging
import sys
import traceback
def get_logger():
    logger = logging.getLogger()
    logger.setLevel(logging.INFO)
    
    handler = logging.StreamHandler()
    formatter = logging.Formatter('%(asctime)s - %(message)s')
    
    handler.setFormatter(formatter)
    logger.addHandler(handler)
    
    return logger

logger = get_logger()
try:
    print(1/0)
except Exception as e:
    logger.exception(sys.exc_info())
    #logger.info(e) / logger.debug(e)
    traceback.print_exc()
```
